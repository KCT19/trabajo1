```python
import cv2  # opencv
import numpy as np
import matplotlib.pyplot as plt

def mostrar_rgb(img, titulo):
    imgrgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.title(titulo)
    plt.imshow(imgrgb)
    plt.axis('off')
    plt.show()

def mostrar_gris(img, titulo):
    img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    plt.title(titulo)
    plt.imshow(img_gray, cmap='gray')
    plt.axis('off')
    plt.show()
    return img_gray
def mostrar_canales(img, titulo):
    plt.rcParams['figure.figsize'] = [10, 5]
    plt.subplot(141)
    plt.title(f'{titulo} - Original')
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    
    plt.subplot(142)
    plt.imshow(img[:, :, 2], cmap='Reds')  
    plt.axis('off')

    plt.subplot(143)
    plt.imshow(img[:, :, 1], cmap='Greens') 
    plt.title('Canal Verde')
    plt.axis('off')

    plt.subplot(144)
    plt.imshow(img[:, :, 0], cmap='Blues')  
    plt.title('Canal Azul')
    plt.axis('off')
    
    plt.show()

imagen1 = cv2.imread('imagen1.jpg')
mostrar_rgb(imagen1, 'Imagen 1 - RGB')
img_gray1 = mostrar_gris(imagen1, 'Imagen 1 - Escala de grises')
cv2.imwrite('imagen1_BGR.png', imagen1)
cv2.imwrite('imagen1_Gris.png', img_gray1)
mostrar_canales(imagen1, 'Imagen 1')

imagen2 = cv2.imread('imagen2.jpg')
mostrar_rgb(imagen2, 'Imagen 2 - RGB')
img_gray2 = mostrar_gris(imagen2, 'Imagen 2 - Escala de grises')
cv2.imwrite('imagen2_BGR.png', imagen2)
cv2.imwrite('imagen2_Gris.png', img_gray2)
mostrar_canales(imagen2, 'Imagen 2')

imagen3 = cv2.imread('imagen3.jpg')
mostrar_rgb(imagen3, 'Imagen 3 - RGB')
img_gray3 = mostrar_gris(imagen3, 'Imagen 3 - Escala de grises')
cv2.imwrite('imagen3_BGR.png', imagen3)
cv2.imwrite('imagen3_Gris.png', img_gray3)
mostrar_canales(imagen3, 'Imagen 3')




```
